```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
```

```python
df2=pd.read_csv('./Heart_Disease_Prediction.csv')
df2.head(20)
```

```python
x=df2.iloc[:,:-1]
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df2['Heart Disease']=label_encoder.fit_transform(df2['Heart Disease'])
y=df2['Heart Disease']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

from sklearn.tree import DecisionTreeClassifier
dectree=DecisionTreeClassifier()
dectree.fit(x_train,y_train)
predDecTree=dectree.predict(x_test)
scoreDecTree=dectree.score(x_test,y_test)
scoreDecTree
```

```python
print(dectree.predict([[64,1,4,128,263,0,0,105,1,0.2,2,1,7]]))
confusion_matrix(y_test, predDecTree)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2['BP'] = scaler.fit_transform(df2[['BP']])
df2['Cholesterol'] = scaler.fit_transform(df2[['Cholesterol']])
df2['Max HR'] = scaler.fit_transform(df2[['Max HR']])
df2
```

```python
x=df2.iloc[:,:-1]
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df2['Heart Disease']=label_encoder.fit_transform(df2['Heart Disease'])
y=df2['Heart Disease']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='lbfgs',max_iter=20000)
lr.fit(x_train,y_train)

predlr=lr.predict(x_test)
score=lr.score(x_test,y_test)
score
```
